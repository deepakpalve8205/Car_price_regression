```python
from scipy.sparse import issparse
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import MinMaxScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer

def Preprocess(x):
    cat=x.select_dtypes(include='object').columns
    con=x.select_dtypes(exclude='object').columns

    num_pipe=Pipeline(steps=[('impute',SimpleImputer(strategy='median')),('scalar',MinMaxScaler())])
    cat_pipe=Pipeline(steps=[('impute',SimpleImputer(strategy='most_frequent')),('Encoder',OneHotEncoder())])

    pre=ColumnTransformer([('Num_pipe',num_pipe,con),('cat_pipe',cat_pipe,cat)])

    x1=pre.fit_transform(x)

    if issparse(x1):
        x1=x1.toarray()
    x2=pd.DataFrame(x1,columns=pre.get_feature_names_out())
    return x2
```
